```python
def find_sentiment(row):
    res = TextBlob(row['review'])
    sen_score = res.sentiment.polarity
    if sen_score >0:
        row['sentiment'] = 'positive'
    else:
        row['sentiment'] = 'negative'
```

```python
df['sentiment'] = ''
```

```python
df.dropna(subset = df.columns,inplace=True)
```

```python
for id,row in tqdm(df.iterrows()):
    #print(id)
    res = TextBlob(row['review'])
    sen_score = res.sentiment.polarity
    if sen_score >0:
        df.loc[id,'sentiment'] = 'positive'
    else:
        df.loc[id,'sentiment'] = 'negative'
```

```python
from hatesonar import Sonar
sonar = Sonar()
```

```python
df['hate_speech']=''
```

```python
for id,row in tqdm(df.iterrows()):
    #print(df.loc[id,'headline'])
    res = sonar.ping(df.loc[id,'review'])
    res = res['classes']
    for item in res:
        if item['class_name'] == 'hate_speech':
                if item['confidence'] > 0.2:
                    df.loc[id,'hate_speech'] = 'hate'
                else:
                    df.loc[id,'hate_speech'] = 'non_hate'
       
```
